```python
data_path = 'fastText_newdata/jbs_skill_augmentations.json'
```

```python
import json

with open(data_path, 'r') as f:
    data_skills = json.load(f)

data_skills
```

```python
from gensim.models import FastText

model_name = 'fastText_newdata/1/fasttext_model'
model = FastText.load(model_name)
```

```python
# set of skill
vocab_skills = []
for skills in data_skills:
    vocab_skills.extend(skills)

vocab_skills
```

```python
def get_frequency_skill(skills_jbs):

    frequency_skills = {}
    for skills in skills_jbs:
        for skill in skills:
            if skill not in list(frequency_skills.keys()):
                frequency_skills[skill] = 0
            else:
                frequency_skills[skill] += 1
    return frequency_skills

def get_topk_frequency(frequency_skills, k=200):
    # Sort the skills by frequency in descending order
    sorted_skills = sorted(frequency_skills.items(), key=lambda x: x[1], reverse=True)
    
    # Get the top K skills
    top_k_skills = sorted_skills[:k]

    # list of tuple to dict
    result_dict = {}

    for item in top_k_skills:
        key, value = item
        result_dict[key] = value
    
    return list(result_dict.keys())
```

```python
frequency_skills = get_frequency_skill(data_skills)
top200 = get_topk_frequency(frequency_skills)
top200
```

```python
# save top200 skills to file top200_skills.json
import json

with open('fastText_newdata/top200_skills.json', 'w') as f:
    json.dump(top200, f, indent=4)
```

```python
def get_topk_sims(skill, model=model, k=5):
    return model.wv.most_similar(skill, topn=k)

def find_nearest_skills(skills):

    result = {}
    
    for skill in skills:
        top5 = get_topk_sims(skill)

        top5 = [x[0] for x in top5]

        result[skill] = top5


    return result
```

```python
relevant_skills = find_nearest_skills(top200)
relevant_skills
```

```python
import json
# save as json
json_path = 'fastText_newdata/1/test_200_skills.json'

with open(json_path, 'w') as f:
    json.dump(relevant_skills, f, indent=4)

print(f"Data saved to {json_path}")
```
